```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
analyser = SentimentIntensityAnalyzer()
```

```python
def sentimentscore(sentence):
    vs = analyser.polarity_scores(sentence)
    print("{:-<65} {}".format(sentence, str(vs)))
```

```python
sentimentscore("This phone is super cool :)")
```

```python
def video(code):
    df=pd.read_csv(f'{code}.csv')
    df.columns = ['user', 'comments', 'date', 'likes']
    df.drop_duplicates()
    df['comments'] = df['comments'].apply(emoji.emojize)
    df['scores'] = df['comments'].apply(lambda x: analyser.polarity_scores(x))
    df['compound']  = df['scores'].apply(lambda score_dict: score_dict['compound'])
    df.to_csv(f'{code}vadered.csv')
    stat= pd.DataFrame({'video':f'{code}',
              'mean':[df['compound'].mean()],
              'median':[df['compound'].mean()],
              'std':[df['compound'].std()],
              '25th':[df['compound'].quantile(q=0.25)],
              '75th':[df['compound'].quantile(q=0.75)],
              'meannoz':[df['compound'][df.compound!=0].mean()],
              'mediannoz':[df['compound'][df.compound!=0].median()],
              'stdnoz':[df['compound'][df.compound!=0].std()],
              '25thnoz':[df['compound'][df.compound!=0].quantile(q=0.25)],
              '75thnoz':[df['compound'][df.compound!=0].quantile(q=0.75)]          
             })
    stat.to_csv(f'{code}stat.csv')
    plt.hist(df['compound'], bins=25, color='xkcd:butter yellow',edgecolor='black', linewidth=1)
    plt.savefig(f'{code}.png')
    plt.clf()
    plt.hist(df['compound'][df.compound!=0], bins=25, color='c',edgecolor='black', linewidth=1)
    plt.savefig(f'{code}noz.png')
    

    
```
